## <span style="color:green;">Step 5: Data Preprocessing</span>

### <span style="color:darkorange;">5.1. Handle Missing Values</span>
- **Replace Missing Values**: Replaces placeholders (e.g., '?') with NaN, then drops rows with missing values to clean the dataset.

### <span style="color:darkorange;">5.2. Remove Duplicates</span>
- **Remove Duplicate Rows**: Ensures dataset integrity by eliminating any duplicate entries.

### <span style="color:darkorange;">5.3. Encode Categorical Variables</span>
- **Label Encoding**: Converts categorical labels into numerical format to be used by machine learning algorithms.

### <span style="color:darkorange;">5.4. Standardize Features</span>
- **Feature Scaling**: Standardizes features to have a mean of 0 and a standard deviation of 1 for improved model performance.

```python
# Data Preprocessing
```

```python
# Handle missing values (assuming '?' represents missing data)
```

```python
df.replace('?', pd.NA, inplace=True)  # Replace '?' with NaN
```

```python
df.dropna(inplace=True)  # Drop rows with missing values
```

```python
# Remove duplicates
```

```python
df.drop_duplicates(inplace=True)  # Remove duplicate rows
```

```python
# Encode categorical variables
```

```python
label_encoder = LabelEncoder()  # Initialize LabelEncoder
```

```python
df['class'] = label_encoder.fit_transform(df['class'])  # Encode class labels
```

```python
# Define features and target variable
```

```python
X = df.drop('class', axis=1)  # Features
```

```python
y = df['class']  # Target variable
```

```python
# Standardize features
```

```python
scaler = StandardScaler()  # Initialize StandardScaler
```

```python
X_scaled = scaler.fit_transform(X)  # Standardize features
```

## <span style="color:green;">Step 6: Train-Test Split</span>

### <span style="color:darkorange;">6.1. Splitting Data</span>
- **Train-Test Split**: Divides the dataset into training and testing subsets, ensuring the model is trained on one subset and evaluated on another to gauge performance.

```python
# Train-test split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)  # Split data into train and test sets
```

## <span style="color:green;">Step 7: Train the Model</span>

### <span style="color:darkorange;">7.1. Initialize and Train Logistic Regression</span>
- **Logistic Regression**: A classification algorithm used to model the probability of a class label based on the input features. The model is trained using the training data.

```python
# Train Logistic Regression model
```

```python
log_reg = LogisticRegression(max_iter=200)  # Initialize Logistic Regression model
```

```python
log_reg.fit(X_train, y_train)  # Train the model
```

```python
# Predict
```

```python
y_pred = log_reg.predict(X_test)  # Predict the test set
```

## <span style="color:green;">Step 8: Model Evaluation</span>

### <span style="color:darkorange;">8.1. Predict and Evaluate</span>
- **Model Prediction**: Uses the trained model to predict class labels on the test set.
- **Accuracy**: Calculates the accuracy of the model on the test set.
- **Confusion Matrix**: Provides a summary of prediction results, showing true positives, true negatives, false positives, and false negatives.

```python
# Evaluate the model
```

```python
accuracy = log_reg.score(X_test, y_test)  # Compute model accuracy
```

```python
print(f"\nModel Accuracy: {accuracy * 100:.2f}%")  # Print accuracy
```

```python
# Convert predictions and true labels to a DataFrame
```

```python
predicted_df = pd.DataFrame({'True Label': y_test.reset_index(drop=True), 'Predicted Label': y_pred})  # Create DataFrame for predictions
```

## <span style="color:green;">Step 9: Visualize Results</span>

### <span style="color:darkorange;">9.1. Display Confusion Matrix</span>
- **Confusion Matrix Plot**: Visualizes the performance of the classification model, showing where the model is making correct and incorrect predictions.

```python
# Display the head of the predicted results
```

```python
print("\nPredicted Results Head:")  # Print header for predicted results head
print(predicted_df.head())  # Display the first few rows of the predicted results
```

```python
# Display info about the predicted results
```

```python
# Confusion Matrix Plot
```

```python
plt.figure(figsize=(8, 6))  # Create a new figure for the confusion matrix plot
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')  # Plot confusion matrix
plt.title('Confusion Matrix')  # Set title of the confusion matrix plot
plt.savefig('confusion_matrix.pdf')  # Save the confusion matrix plot as a PDF file
plt.show()  # Display the confusion matrix plot
```
