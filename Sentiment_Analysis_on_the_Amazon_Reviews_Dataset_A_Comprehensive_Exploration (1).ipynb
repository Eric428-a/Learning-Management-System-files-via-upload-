### <span style="color:orange">Step 6: Load and Process Data</span>

- **Objective**:
  - Load the Amazon Reviews dataset into memory.
  - Preprocess the data for further analysis and model training.

- **Tasks**:
  1. **Load Compressed Files**:
     - Define paths to the compressed files (`train.ft.txt.bz2` and `test.ft.txt.bz2`) within the extracted dataset directory.
     - Implement a function to decompress these files and retrieve the raw text data.

  2. **Split Labels and Reviews**:
     - Parse the raw data to separate labels and corresponding review texts.
     - Handle any formatting or encoding issues during data extraction.

  3. **Create DataFrames**:
     - Construct pandas DataFrames (`train_df` and `test_df`) from the parsed data.
     - Ensure proper column naming (`label` and `review`) for structured representation.

- **Implementation Details**:
  - **Python Imports**:
    - `import bz2`: Utilized for decompressing bz2 files.
    - `import pandas as pd`: Essential for creating and manipulating DataFrames.
  
  - **Code Example**:
    ```python
    train_file_path = os.path.join(dataset_dir, 'train.ft.txt.bz2')
    test_file_path = os.path.join(dataset_dir, 'test.ft.txt.bz2')

    def extract_bz2(file_path):
        with open(file_path, 'rb') as f_in:
            data = bz2.decompress(f_in.read()).decode('utf-8')
        return data

    train_data = extract_bz2(train_file_path)[:50000]  # Sample: Load first 50,000 characters
    test_data = extract_bz2(test_file_path)[:50000]    # Sample: Load first 50,000 characters

    def split_lines(data):
        lines = data.splitlines()
        labels = [line.split(' ')[0] for line in lines]
        reviews = [' '.join(line.split(' ')[1:]) for line in lines]
        return labels, reviews

    train_labels, train_reviews = split_lines(train_data)
    test_labels, test_reviews = split_lines(test_data)

    train_df = pd.DataFrame({'label': train_labels, 'review': train_reviews}).head(1000)
    test_df = pd.DataFrame({'label': test_labels, 'review': test_reviews}).head(1000)
    ```
  
- **Notes**:
  - Adjust the `[:50000]` slicing for `train_data` and `test_data` as needed to handle larger datasets or specific data sampling requirements.
  - Ensure data integrity and proper handling of text encoding and formatting during the extraction and DataFrame creation stages.
  - Verify the structure and content of `train_df` and `test_df` DataFrames to confirm accurate representation of labels and review texts before proceeding with further analysis or model training.

````python
train_file_path = os.path.join(dataset_dir, 'train.ft.txt.bz2')
````

````python
test_file_path = os.path.join(dataset_dir, 'test.ft.txt.bz2')
````

````python
train_data = extract_bz2(train_file_path)[:50000]  # Load first 50,000 characters for demo
````

````python
test_data = extract_bz2(test_file_path)[:50000]    # Load first 50,000 characters for demo
````

````python
# Split lines into label and review
````

````python
def split_lines(data):
    lines = data.splitlines()  # Split data into lines
    labels = [line.split(' ')[0] for line in lines]  # Extract labels
    reviews = [' '.join(line.split(' ')[1:]) for line in lines]  # Extract reviews
    return labels, reviews
````

````python
train_labels, train_reviews = split_lines(train_data)
````

````python
test_labels, test_reviews = split_lines(test_data)
````

````python
# Create DataFrames
````

````python
train_df = pd.DataFrame({'label': train_labels, 'review': train_reviews}).head(1000)  # Create train DataFrame
````

````python
test_df = pd.DataFrame({'label': test_labels, 'review': test_reviews}).head(1000)  # Create test DataFrame
````

###Step 8: Preprocess the Data

- **<span style="color:blue">Objective</span>**:
  - The objective of this step is to prepare the Amazon Reviews dataset for training a BERT-based sentiment classification model. This involves tokenizing the textual reviews using a BERT tokenizer and converting labels into a suitable format for model training.

- **<span style="color:blue">Tasks</span>**:
  1. **<span style="color:green">Tokenization</span>**:
     - **<span style="color:purple">Method</span>**: Utilize the BERT tokenizer (`BertTokenizer.from_pretrained('bert-base-uncased')`) to tokenize the reviews.
     - **<span style="color:purple">Explanation</span>**: Tokenization converts textual data into numerical tokens that can be processed by machine learning models. Parameters such as `padding=True`, `truncation=True`, and `max_length=128` ensure uniformity in sequence length and efficient computation.

  2. **<span style="color:green">Label Conversion</span>**:
     - **<span style="color:purple">Method</span>**: Convert sentiment labels ('__label__1' for negative and '__label__2' for positive) into integer format (`train_labels` and `test_labels`).
     - **<span style="color:purple">Explanation</span>**: Model training requires numeric labels. Converting '__label__1' to 0 and '__label__2' to 1 aligns the labels with the binary classification task for sentiment analysis.

- **<span style="color:blue">Implementation Details</span>**:
  - **<span style="color:green">Python Functions</span>**:
    - `tokenize_reviews(reviews, tokenizer, max_length=128)`: A function that applies tokenization to a list of reviews using the specified tokenizer and parameters.
    - `BertTokenizer.from_pretrained('bert-base-uncased')`: Initializes the BERT tokenizer from the Hugging Face transformers library, which is pretrained on uncased English text.

  - **<span style="color:green">Data Transformation</span>**:
    - `train_tokenized` and `test_tokenized`: Tokenized representations of the training and test reviews, respectively, suitable for input into a BERT-based model.
    - `train_labels` and `test_labels`: Integer representations of sentiment labels ('0' for negative and '1' for positive) for the training and test sets.

- **<span style="color:blue">Considerations</span>**:
  - **<span style="color:green">Data Integrity</span>**: Ensure that tokenization and label conversion accurately reflect the original dataset structure to maintain data integrity.
  - **<span style="color:green">Model Compatibility</span>**: Verify that the tokenized data format aligns with the input requirements of the BERT-based sentiment classification model.

````python
# Preprocess the data
````

````python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')  # Initialize BERT tokenizer
````

````python
def tokenize_reviews(reviews, tokenizer, max_length=128):
    tokenized = tokenizer(reviews, padding=True, truncation=True, max_length=max_length, return_tensors='pt')  # Tokenize reviews
    return tokenized
````

````python
train_tokenized = tokenize_reviews(train_df['review'].tolist(), tokenizer)  # Tokenizing train reviews
````

````python
test_tokenized = tokenize_reviews(test_df['review'].tolist(), tokenizer)  # Tokenizing test reviews
````

````python
# Convert labels to integers
````

````python
train_labels = [int(label.split('__label__')[1]) - 1 for label in train_df['label']]  # Converting train labels to integers
````

````python
test_labels = [int(label.split('__label__')[1]) - 1 for label in test_df['label']]  # Converting test labels to integers
````

###Step 9: Define a Custom Dataset Class

- **<span style="color:blue">Objective</span>**:
  - The objective of this step is to create a custom dataset class (`AmazonReviewsDataset`) in PyTorch. This class will facilitate the loading and processing of the Amazon Reviews dataset, tokenized using BERT's tokenizer, and formatted for efficient training and evaluation of a sentiment classification model.

- **<span style="color:blue">Tasks</span>**:
  1. **<span style="color:green">Class Definition</span>**:
     - **<span style="color:purple">Implementation</span>**: Define a class `AmazonReviewsDataset` that inherits from `torch.utils.data.Dataset`.
     - **<span style="color:purple">Constructor (`__init__`)</span>**: Initialize the class with `encodings` and `labels`. These parameters represent the tokenized input data and corresponding sentiment labels, respectively.
     - **<span style="color:purple">`__getitem__` Method</span>**: Implement this method to return a dictionary (`item`) containing tokenized encodings (`encodings`) and their corresponding labels (`labels`) for a specific index (`idx`). This method enables efficient retrieval and batching of data during training and evaluation.
     - **<span style="color:purple">`__len__` Method</span>**: Define this method to return the total number of items in the dataset, which is determined by the length of `labels`. This ensures that the dataset class behaves correctly with PyTorch's data loaders (`DataLoader`) for iterating over batches of data.

  2. **<span style="color:green">Dataset Instances</span>**:
     - **<span style="color:purple">Initialization</span>**: Instantiate `train_dataset` and `test_dataset` using the `AmazonReviewsDataset` class.
     - **<span style="color:purple">Parameters</span>**: Initialize `train_dataset` with `train_tokenized` (tokenized reviews for training) and `train_labels` (corresponding sentiment labels). Similarly, initialize `test_dataset` with `test_tokenized` (tokenized reviews for testing) and `test_labels` (corresponding sentiment labels).

- **<span style="color:blue">Implementation Details</span>**:
  - **<span style="color:green">Purpose</span>**: Custom dataset classes (`Dataset`) in PyTorch are essential for handling specialized data formats, such as tokenized text data for natural language processing tasks like sentiment analysis.
  - **<span style="color:green">Tensor Conversion</span>**: Use `torch.tensor` to convert data into PyTorch tensors within the `__getitem__` method. This conversion ensures that the data is compatible with GPU acceleration and deep learning models, enhancing training efficiency.
  - **<span style="color:green">Efficient Data Management</span>**: By implementing a custom dataset class, you streamline the process of loading, processing, and feeding data into your model. This approach optimizes memory usage and facilitates seamless integration with PyTorch's training pipelines.

- **<span style="color:blue">Considerations</span>**:
  - **<span style="color:green">Data Integrity</span>**: Ensure that `train_tokenized`, `test_tokenized`, `train_labels`, and `test_labels` are correctly aligned with their respective dataset splits. This alignment prevents data leakage and maintains the integrity of model validation during training.
  - **<span style="color:green">Performance Optimization</span>**: Verify that the custom dataset class efficiently manages computational resources, especially when dealing with large-scale datasets or constrained hardware environments. Consider strategies like batch loading and prefetching to further enhance performance.


````python
# Define a custom dataset class
````

````python
class AmazonReviewsDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings  # Storing tokenized encodings
        self.labels = labels  # Storing labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}  # Getting item by index
        item['labels'] = torch.tensor(self.labels[idx])  # Getting labels
        return item

    def __len__(self):
        return len(self.labels)  # Returning length of labels
````

````python
train_dataset = AmazonReviewsDataset(train_tokenized, train_labels)  # Creating train dataset
````

````python
test_dataset = AmazonReviewsDataset(test_tokenized, test_labels)  # Creating test dataset
````
