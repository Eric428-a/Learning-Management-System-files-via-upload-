# <span style="color:blue">**Step 5: Handle Missing Values**</span>

In this optional step, we handle the missing values:

- **`df.fillna()`**: This function allows us to replace missing values. In this case, we are filling missing values for numeric columns with the mean values using the `inplace=True` parameter to modify the DataFrame directly.

```python
# Step 5: Handle Missing Values (optional step)
# Depending on our analysis, we might decide to fill in missing values or drop rows with missing values.
```

For numerical columns, we use **`histplot()`** with kernel density estimation (KDE) to visualize the distribution of the data.

# <span style="color:blue">**Step 11: Handle Categorical Variables**</span>

Categorical variables need to be encoded as numerical values for most machine learning algorithms. In this step, we use **`pd.get_dummies()`** to convert categorical variables into dummy/indicator variables. This process transforms categories into binary columns, making them suitable for analysis while avoiding the introduction of bias due to numerical ordering.

```python
# Step 11: Handle Categorical Variables (optional step)
# If our data contains categorical variables, we might need to encode them into numerical values.
```

```python
df = pd.get_dummies(df, drop_first=True)  # Convert categorical variables into dummy/indicator variables.
```

# <span style="color:blue">**Step 12: Outlier Detection**</span>

Outliers can distort the results of our analysis. In this step, we detect and remove outliers using the **Interquartile Range (IQR)** method. The IQR represents the middle 50% of the data, and values outside the range [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] are considered outliers. Removing these values helps prevent skewed results.

```python
# Step 12: Outlier Detection (optional step)
# Outliers can significantly affect our analysis, so we use methods like IQR score or Z-score to detect and handle them.
```

```python
Q1 = df.quantile(0.25)
```

```python
Q3 = df.quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]  # Remove outliers.
```

# <span style="color:blue">**Step 13: Feature Scaling**</span>

In this step, we scale the features using **`StandardScaler()`**, which standardizes the data by removing the mean and scaling to unit variance. This is especially important when features have different ranges, as it ensures that all features contribute equally to the model. Standardization is a crucial preprocessing step for many machine learning algorithms.

```python
# Step 13: Feature Scaling (optional step)
# If our dataset contains features with different scales, we might need to scale them to a standard scale.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)  # Scale the features to a standard scale.
```
