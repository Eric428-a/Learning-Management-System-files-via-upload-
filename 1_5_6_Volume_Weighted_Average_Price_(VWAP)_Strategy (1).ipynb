## <span style="color: #d62728;">3. Feature Engineering</span>

### 3.1 Typical Price

- **Definition**: The Typical Price represents a single value that averages the high, low, and close prices for each period, providing a more balanced view of the price level.
- **Formula**: \[(High + Low + Close) / 3\]
- **Purpose**: This calculation smoothens price data and provides a representative price point for each period, which is crucial for the VWAP calculation.

### 3.2 VWAP

- **Definition**: VWAP is a cumulative measure that combines price and volume to reflect the average price at which a security has traded over a specific period, typically one day.
- **Formula**:
  \[
  VWAP = \frac{ \sum_{i=1}^{n} (Typical Price_i \times Volume_i) }{ \sum_{i=1}^{n} Volume_i }
  \]
  where \( Typical Price_i \) is the typical price for the \( i^{th} \) period and \( Volume_i \) is the trading volume for that period.
- **Purpose**: VWAP provides a benchmark for evaluating trading performance. It helps traders identify trends and potential entry or exit points by comparing the current price to the VWAP.

## <span style="color: #2ca02c;">1.1 Understanding VWAP in Machine Learning Context</span>

### 1.1.1 What is VWAP?

- **Definition**: VWAP is the weighted average price of a security, calculated by taking the cumulative sum of the product of the typical price and volume, divided by the cumulative volume. It provides a measure of the average price a security has traded at throughout the day, adjusted for volume.
- **Formula**:
  \[
  VWAP_t = \frac{ \sum_{i=1}^{t} (Typical Price_i \times Volume_i) }{ \sum_{i=1}^{t} Volume_i }
  \]
  where \( Typical Price_i \) is the typical price for the \( i^{th} \) period, and \( Volume_i \) is the trading volume for that period.

### 1.1.2 Importance of VWAP in Trading

- **Benchmarking**: VWAP is used as a benchmark to assess the quality of trades. If a trader’s execution price is below VWAP, it suggests they bought below the average price of the day; if above, they sold above the average.
- **Trading Decisions**: VWAP helps in making buy or sell decisions. Prices above VWAP might indicate an overvalued security, while prices below VWAP might suggest undervaluation.

# <span style="color: #d62728;">3. Feature Engineering</span>

## <span style="color: #ff7f0e;">3.1 Typical Price</span>

### 3.1.1 Definition

- **Typical Price**: Represents a single value that averages the high, low, and close prices for each period.
- **Purpose**: Provides a balanced view of the price level for each period, smoothing out extreme values.

### 3.1.2 Calculation

- **Formula**:
  \[
  \text{Typical Price} = \frac{ \text{High} + \text{Low} + \text{Close} }{3}
  \]
- **Example Calculation**:
  - If `High = $50`, `Low = $45`, and `Close = $48`, then:
    \[
    \text{Typical Price} = \frac{50 + 45 + 48}{3} = 47.67
    \]

### 3.1.3 Importance

- **Role in VWAP**: Serves as a key input in the VWAP calculation by providing a representative price for each period.

## <span style="color: #ff7f0e;">3.2 VWAP (Volume-Weighted Average Price)</span>

### 3.2.1 Definition

- **VWAP**: A cumulative measure that reflects the average price at which a security has traded over a specific period, adjusted for volume.
- **Purpose**: Provides a benchmark for evaluating trading performance and trends.

### 3.2.2 Calculation

- **Formula**:
  \[
  \text{VWAP} = \frac{ \sum_{i=1}^{n} (\text{Typical Price}_i \times \text{Volume}_i) }{ \sum_{i=1}^{n} \text{Volume}_i }
  \]
- **Components**:
  - **Typical Price**: Calculated as above.
  - **Volume**: Trading volume for the period.
- **Example Calculation**:
  - Assume Typical Prices and Volumes for three periods are as follows:
    - Period 1: Typical Price = $47.67, Volume = 100
    - Period 2: Typical Price = $48.50, Volume = 150
    - Period 3: Typical Price = $46.75, Volume = 200
  - VWAP Calculation:
    \[
    \text{VWAP} = \frac{(47.67 \times 100) + (48.50 \times 150) + (46.75 \times 200)}{100 + 150 + 200} = \frac{4767 + 7275 + 9350}{450} = 47.87
    \]

### 3.2.3 Importance

- **Role**: Helps traders identify trends and potential entry or exit points by comparing the current price to the VWAP. A price above VWAP might indicate overvaluation, while a price below VWAP might suggest undervaluation.

## <span style="color: #ff7f0e;">Overview</span>

The **VWAP and Bollinger Bands** graph displays the Volume-Weighted Average Price (VWAP) alongside Bollinger Bands. This visualization helps in analyzing how VWAP aligns with market prices and assessing price volatility.

## <span style="color: #2ca02c;">Purpose</span>

- **VWAP Tracking**: Compares VWAP with actual closing prices to evaluate how VWAP reflects the average trading price.
- **Volatility Analysis**: Bollinger Bands provide insight into market volatility and potential buy/sell signals.
- **Trading Signals**: Helps in identifying trends, potential entry, and exit points based on price deviations from the VWAP and Bollinger Bands.

## <span style="color: #d62728;">Key Features</span>

- **VWAP Line**: Shows the average price weighted by volume.
- **Bollinger Bands**:
  - **Middle Band**: Typically a 20-day moving average.
  - **Upper Band**: Middle Band + (2 × standard deviation).
  - **Lower Band**: Middle Band - (2 × standard deviation).
- **X-Axis**: Represents time.
- **Y-Axis**: Represents price and Bollinger Bands.

## <span style="color: #9467bd;">Interpretation</span>

- **Price Above VWAP**: Indicates that the current price is higher than the average price, potentially signaling overvaluation.
- **Price Below VWAP**: Suggests that the current price is lower than the average price, possibly indicating undervaluation.
- **Bollinger Bands Width**: A wider band indicates higher volatility, while a narrower band suggests lower volatility.

````python
# Feature Engineering
````

````python
# Calculate Typical Price
````

````python
data['Typical Price'] = (data['High'] + data['Low'] + data['Close']) / 3  # Compute the typical price
````

````python
# Calculate VWAP
````

````python
data['VWAP'] = (data['Typical Price'] * data['Volume_(BTC)']).cumsum() / data['Volume_(BTC)'].cumsum()  # Compute VWAP using cumulative sums
````

````python
# Plot VWAP
````

````python
plt.figure(figsize=(14, 7))  # Set the figure size
plt.plot(data.index, data['Close'], label='Close Price', color='blue', alpha=0.5)  # Plot Close price
plt.plot(data.index, data['VWAP'], label='VWAP', color='orange')  # Plot VWAP
plt.title('Volume-Weighted Average Price (VWAP) Strategy')  # Set the title of the plot
plt.xlabel('Date')  # Set x-axis label
plt.ylabel('Price')  # Set y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines to the plot
plt.savefig('/content/vwap_strategy.png')  # Save the plot as a PNG file
plt.show()  # Display the plot
````

````python
# Calculate Bollinger Bands
````

````python
data['20-MA'] = data['Close'].rolling(window=20).mean()  # Compute 20-period moving average
````

````python
data['20-STD'] = data['Close'].rolling(window=20).std()  # Compute 20-period standard deviation
````

````python
data['Upper Band'] = data['20-MA'] + (data['20-STD'] * 2)  # Compute Upper Bollinger Band
````

````python
data['Lower Band'] = data['20-MA'] - (data['20-STD'] * 2)  # Compute Lower Bollinger Band
````

````python
# Plot Bollinger Bands
````

````python
plt.figure(figsize=(14, 7))  # Set the figure size
plt.plot(data.index, data['Close'], label='Close Price', color='blue', alpha=0.5)  # Plot Close price
plt.plot(data.index, data['20-MA'], label='20-Day Moving Average', color='green')  # Plot 20-period moving average
plt.plot(data.index, data['Upper Band'], label='Upper Bollinger Band', color='red')  # Plot Upper Bollinger Band
plt.plot(data.index, data['Lower Band'], label='Lower Bollinger Band', color='red')  # Plot Lower Bollinger Band
plt.fill_between(data.index, data['Lower Band'], data['Upper Band'], color='red', alpha=0.1)  # Fill between Upper and Lower Bollinger Bands
plt.title('Bollinger Bands')  # Set the title of the plot
plt.xlabel('Date')  # Set x-axis label
plt.ylabel('Price')  # Set y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines to the plot
plt.savefig('/content/bollinger_bands.png')  # Save the plot as a PNG file
plt.show()  # Display the plot
````

## <span style="color: #1f77b4;">5. VWAP Strategy Implementation</span>

### <span style="color: #ff7f0e;">Objective:</span>
Implement the VWAP strategy by generating trading signals based on the relationship between the VWAP and the closing price. This step is crucial for identifying potential trading opportunities and informing trading decisions.

### <span style="color: #2ca02c;">Generating Trading Signals</span>

#### <span style="color: #d62728;">Buy Signal:</span>
- **Definition**: A buy signal is generated when the closing price is below the VWAP. This indicates that the security is undervalued compared to its average trading price, suggesting a potential buying opportunity.
- **Implementation**:
  - Create a new column in the DataFrame to denote buy signals.
  - **Code Snippet**:
    ```python
    # Generate Buy Signal
    df_preprocessed['Buy_Signal'] = df_preprocessed['Close'] < df_preprocessed['VWAP']
    ```
  - **Explanation**: This code creates a boolean column, `Buy_Signal`, where `True` indicates that the closing price is below the VWAP, signaling a potential buy.

#### <span style="color: #9467bd;">Sell Signal:</span>
- **Definition**: A sell signal is generated when the closing price is above the VWAP. This suggests that the security is overvalued compared to its average trading price, indicating a potential selling opportunity.
- **Implementation**:
  - Create a new column in the DataFrame to denote sell signals.
  - **Code Snippet**:
    ```python
    # Generate Sell Signal
    df_preprocessed['Sell_Signal'] = df_preprocessed['Close'] > df_preprocessed['VWAP']
    ```
  - **Explanation**: This code creates a boolean column, `Sell_Signal`, where `True` indicates that the closing price is above the VWAP, signaling a potential sell.

### <span style="color: #e377c2;">Purpose of Signals</span>

- **Buy Signal**:
  - **Objective**: Identify potential buying opportunities when the market price is lower than the average trading price, with the expectation of future price appreciation.
  - **Use Case**: Traders might enter positions when they see a buy signal, anticipating that the price will rise towards or above the VWAP.

- **Sell Signal**:
  - **Objective**: Identify potential selling opportunities when the market price is higher than the average trading price, with the expectation of future price depreciation.
  - **Use Case**: Traders might exit positions or short-sell when they see a sell signal, anticipating that the price will fall towards or below the VWAP.

### <span style="color: #8c564b;">Visualizing Signals</span>

#### <span style="color: #7f7f7f;">Buy and Sell Signals Visualization:</span>
- **Objective**: To visualize the buy and sell signals alongside VWAP and closing prices to understand their effectiveness and alignment with market movements.
- **Implementation**:
  - Plot closing prices, VWAP, and the generated buy/sell signals on a graph to visually assess trading opportunities.
  - **Code Snippet**:
    ```python
    import matplotlib.pyplot as plt

    plt.figure(figsize=(14, 7))

    # Plot closing price
    plt.plot(df_preprocessed['Timestamp'], df_preprocessed['Close'], label='Close Price', color='blue')

    # Plot VWAP
    plt.plot(df_preprocessed['Timestamp'], df_preprocessed['VWAP'], label='VWAP', color='red')

    # Plot Buy Signals
    buy_signals = df_preprocessed[df_preprocessed['Buy_Signal']]
    plt.scatter(buy_signals['Timestamp'], buy_signals['Close'], marker='^', color='green', label='Buy Signal', alpha=1)

    sell_signals = df_preprocessed[df_preprocessed['Sell_Signal']]
    plt.scatter(sell_signals['Timestamp'], sell_signals['Close'], marker='v', color='red', label='Sell Signal', alpha=1)

    plt.title('VWAP Strategy - Buy and Sell Signals')
    plt.xlabel('Timestamp')
    ```

  - **Explanation**: This code generates a plot showing the closing prices, VWAP, and the buy/sell signals. Buy signals are marked with green upward triangles and sell signals with red downward triangles.

### <span style="color: #1f77b4;">Summary:</span>
In Step 5, the VWAP strategy is implemented by generating and visualizing buy and sell signals based on the comparison between closing prices and VWAP. This helps traders identify potential trading opportunities and evaluate the strategy's effectiveness.

````python
#  VWAP Strategy Implementation
````

````python
# Create signals: Buy when Close < VWAP, Sell when Close > VWAP
````

````python
scaled_data['Signal'] = 0  # Initialize the 'Signal' column with zeros
````

````python
scaled_data['Signal'][scaled_data['Close'] < scaled_data['VWAP']] = 1  # Set Buy signal (1) when Close < VWAP
````

````python
scaled_data['Signal'][scaled_data['Close'] > scaled_data['VWAP']] = -1  # Set Sell signal (-1) when Close > VWAP
````

## <span style="color: #1f77b4;">6. Backtesting the Strategy</span>

### <span style="color: #ff7f0e;">Objective:</span>
Backtesting evaluates the performance of the VWAP strategy by simulating trades using historical data. This helps in understanding how well the strategy would have performed in the past and provides insights into its effectiveness.

### <span style="color: #2ca02c;">Computing Returns</span>

#### <span style="color: #d62728;">Daily Returns:</span>
- **Definition**: Daily returns are the percentage change in the price of the security from one day to the next. This provides insight into the daily fluctuations in price.
- **Implementation**: Calculate daily returns to understand the price movement on a daily basis.

#### <span style="color: #9467bd;">Strategy Returns:</span>
- **Definition**: Strategy returns reflect the performance of the VWAP strategy based on the buy and sell signals generated.
- **Implementation**: Calculate returns based on these signals to determine the effectiveness of the strategy.

#### <span style="color: #e377c2;">Cumulative Returns:</span>
- **Definition**: Cumulative returns show the total return over time, considering all individual daily returns.
- **Implementation**: Compute cumulative returns to assess the overall performance of the strategy over the entire period.

### <span style="color: #7f7f7f;">Performance Evaluation</span>

#### <span style="color: #8c564b;">Comparing Strategy to Benchmark:</span>
- **Objective**: Compare the VWAP strategy's cumulative returns with a benchmark (such as a buy-and-hold strategy) to evaluate its performance.
- **Implementation**: Analyze how the strategy performs relative to a simple benchmark to gauge its effectiveness.

#### <span style="color: #1f77b4;">Visualization:</span>
- **Objective**: Visualize the cumulative returns of the VWAP strategy and benchmark to assess the strategy's performance visually.
- **Implementation**: Plot the cumulative returns to compare the strategy against the benchmark and evaluate its success.

````python
# Backtesting the Strategy
````

````python
# Compute returns
````

````python
scaled_data['Returns'] = scaled_data['Close'].pct_change()  # Compute daily returns
````

````python
scaled_data['Strategy Returns'] = scaled_data['Returns'] * scaled_data['Signal'].shift(1)  # Compute strategy returns based on signals
````

````python
scaled_data['Cumulative Returns'] = (1 + scaled_data['Strategy Returns']).cumprod()  # Compute cumulative returns of the strategy
````

## <span style="color: #d62728;">Key Features</span>

- **Rolling Mean**: Shows the average VWAP over a rolling window.
- **Rolling Standard Deviation**: Displays the variability of VWAP over a rolling window.
- **X-Axis**: Represents time.
- **Y-Axis**: Represents VWAP values and statistics.

## <span style="color: #9467bd;">Interpretation</span>

- **Mean Trends**: Observing trends in the rolling mean helps in understanding the direction of VWAP over time.
- **Volatility Insights**: High rolling standard deviation indicates periods of high volatility, while low deviation suggests stability.
- **Strategy Adjustments**: Provides information for fine-tuning trading strategies based on VWAP’s volatility and trends.

````python
# Additional Visualization: Rolling Mean and Standard Deviation of VWAP
````

````python
rolling_window = 30  # Define the rolling window size
````

````python
data['VWAP Rolling Mean'] = data['VWAP'].rolling(window=rolling_window).mean()  # Compute rolling mean of VWAP
````

````python
data['VWAP Rolling Std'] = data['VWAP'].rolling(window=rolling_window).std()  # Compute rolling standard deviation of VWAP
````

````python
plt.figure(figsize=(14, 7))  # Set the figure size
plt.plot(data.index, data['VWAP'], label='VWAP', color='orange', alpha=0.7)  # Plot VWAP
plt.plot(data.index, data['VWAP Rolling Mean'], label=f'{rolling_window}-Minute Rolling Mean', color='blue')  # Plot rolling mean
plt.fill_between(data.index, data['VWAP Rolling Mean'] - data['VWAP Rolling Std'], data['VWAP Rolling Mean'] + data['VWAP Rolling Std'], color='blue', alpha=0.1, label=f'{rolling_window}-Minute Rolling Std Dev')  # Fill between rolling mean ± rolling std dev
plt.title('VWAP with Rolling Mean and Std Dev')  # Set the title of the plot
plt.xlabel('Date')  # Set x-axis label
plt.ylabel('VWAP')  # Set y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines to the plot
plt.savefig('/content/vwap_rolling_mean_std.png')  # Save the plot as a PNG file
plt.show()  # Display the plot
````
