## <font color='green'>**Step 3: Portfolio Optimization Functions**</font>

### **Portfolio Performance Calculation**
- **Define Function**: Create a function to calculate portfolio performance metrics such as return and standard deviation.
  - Inputs: Weights of assets, mean returns, and covariance matrix.
  - Outputs: Portfolio return and standard deviation.

### **Objective Function**
- **Maximize Utility**: Define an objective function to maximize portfolio utility, typically the Sharpe ratio.
  - Utility Calculation: Calculate utility as `(portfolio_return - risk_free_rate) / portfolio_std_dev`.
  - Optimization Goal: Minimize negative utility to maximize Sharpe ratio.

### **Optimize Portfolio**
- **Initial Weights**: Set initial weights for assets, usually equally distributed.
- **Constraints**: Ensure that the sum of weights equals 1, representing a fully invested portfolio.
- **Bounds**: Set bounds for weights between 0 and 1, ensuring no short selling.
- **Optimization**: Utilize optimization techniques like `scipy.optimize.minimize` to find optimal weights that maximize the Sharpe ratio.

### **Next Steps**
After defining portfolio optimization functions, proceed to Step 4 where you will apply these functions to calculate mean returns, covariance matrices, and optimize portfolio weights based on the data processed in Step 2.

```python
# Step 3: Portfolio Optimization Functions
```

```python
def calculate_portfolio_performance(weights, mean_returns, cov_matrix):
    portfolio_return = np.sum(mean_returns * weights)  # Calculate portfolio return
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))  # Calculate portfolio standard deviation
    return portfolio_return, portfolio_std_dev  # Return portfolio return and standard deviation
```

```python
def objective_function(weights, mean_returns, cov_matrix, risk_free_rate):
    portfolio_return, portfolio_std_dev = calculate_portfolio_performance(weights, mean_returns, cov_matrix)
    utility = (portfolio_return - risk_free_rate) / portfolio_std_dev  # Calculate portfolio utility (Sharpe ratio)
    return -utility  # Maximize utility is equivalent to minimizing negative utility
```

```python
def optimize_portfolio(mean_returns, cov_matrix, risk_free_rate):
    num_assets = len(mean_returns)  # Number of assets
    initial_weights = np.array([1 / num_assets] * num_assets)  # Initial equal weights for optimization
    constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})  # Sum of weights constraint
    bounds = tuple((0, 1) for _ in range(num_assets))  # Bounds for weights
    result = minimize(objective_function, initial_weights, args=(mean_returns, cov_matrix, risk_free_rate),
                      method='SLSQP', bounds=bounds, constraints=constraints)  # Minimize objective function
    return result.x  # Return optimal weights
```

## <font color='green'>**Step 9: Evaluate Portfolio Performance Metrics**</font>

### **Sharpe Ratio Calculation**
- **Definition**: Calculate the Sharpe ratio to evaluate risk-adjusted return.
- **Formula**: \( Sharpe\ ratio = \frac{{R_p - R_f}}{{\sigma_p}} \), where \( R_p \) is the portfolio return, \( R_f \) is the risk-free rate, and \( \sigma_p \) is the portfolio standard deviation.
- **Implementation**: Compute the Sharpe ratio using historical returns and standard deviation.

### **Sortino Ratio Calculation**
- **Purpose**: Assess downside risk-adjusted return using the Sortino ratio.
- **Formula**: \( Sortino\ ratio = \frac{{R_p - R_f}}{{\sigma_d}} \), where \( \sigma_d \) is the downside deviation (standard deviation of negative returns).
- **Implementation**: Calculate the Sortino ratio to complement the Sharpe ratio in risk assessment.

- **Volatility Measurement**: Determine the portfolio's volatility or standard deviation of returns as a measure of risk.
  - Example: Use the formula \( \sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (R_i - \bar{R})^2} \), where \( R_i \) is the return for period \( i \) and \( \bar{R} \) is the mean return.
- **Sharpe Ratio Calculation**: Evaluate risk-adjusted performance using the Sharpe ratio, which compares the portfolio's excess return to its volatility.
  - Example: \( Sharpe Ratio = \frac{R_p - R_f}{\sigma_p} \), where \( R_p \) is the portfolio return, \( R_f \) is the risk-free rate, and \( \sigma_p \) is the portfolio standard deviation.

### **Benchmark Comparison**
- **Select Benchmark**: Choose an appropriate benchmark index or portfolio against which to compare performance.
- **Performance Differential**: Calculate the excess return (alpha) of the portfolio relative to the benchmark.
  - Example: \( \text{Alpha} = R_p - R_b \), where \( R_b \) is the benchmark return.

### **Risk-Adjusted Metrics**
- **Sortino Ratio**: Measure downside risk-adjusted return using the Sortino ratio, which focuses on negative deviations from the mean return.
  - Example: \( \text{Sortino Ratio} = \frac{R_p - R_f}{\sigma_{downside}} \), where \( \sigma_{downside} \) is the standard deviation of negative returns.
