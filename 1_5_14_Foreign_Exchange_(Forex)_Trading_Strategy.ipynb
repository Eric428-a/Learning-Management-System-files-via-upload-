## **<span style="color:green">Data Preparation</span>**

### **<span style="color:green">Overview</span>**
Data preparation is a crucial step in any data analysis project. This phase involves cleaning and transforming raw data into a suitable format for analysis. Proper data preparation ensures that the dataset is ready for feature engineering, model training, and evaluation. This step includes handling missing values, encoding categorical variables, scaling numerical features, and removing or imputing outliers.

### **<span style="color:green">Steps and Notes</span>**

1. **Handling Missing Values**
   - **Objective**: Ensure that the dataset has no missing values which can adversely affect the analysis.
   - **Methods**:
     - If missing values exist, decide on a strategy to handle them (e.g., filling with mean/median, forward/backward filling, or dropping rows/columns).
   - **Notes**:
     - For this dataset, there were no missing values, simplifying this step.

2. **Handling Duplicates**
   - **Objective**: Remove duplicate rows to maintain data integrity.
   - **Methods**:
     - Use `.drop_duplicates()` to remove any duplicate entries in the dataset.
   - **Notes**:
     - Confirm the removal of duplicates by re-checking the dataset.

3. **Feature Engineering**
   - **Objective**: Create new features or modify existing ones to enhance the predictive power of the model.
   - **Methods**:
     - Example: Creating technical indicators for forex data, such as Moving Averages, RSI, etc.
   - **Notes**:
     - Carefully select and engineer features that are relevant to the trading strategy.

4. **Scaling Numerical Features**
   - **Objective**: Normalize numerical features to ensure that they contribute equally to the analysis.
   - **Methods**:
     - Use standardization (mean=0, std=1) or normalization (scaling between 0 and 1) techniques.
   - **Notes**:
     - Scaling is essential for algorithms that rely on distance metrics, such as K-Nearest Neighbors and SVM.

5. **Handling Outliers**
   - **Objective**: Identify and appropriately handle outliers to prevent them from skewing the analysis.
   - **Methods**:
     - Use visualization techniques such as box plots to identify outliers.
     - Decide on a strategy to handle outliers (e.g., removal, transformation, or capping).
   - **Notes**:
     - Outliers can provide important insights but can also distort the results if not handled properly.

- **Feature Relevance**: Only include features that add value to the analysis. Irrelevant features can introduce noise.

### **<span style="color:blue">Feature Engineering</span>**
Feature engineering is a critical phase in a data science project that involves creating new features or modifying existing ones to enhance the performance of machine learning models. In this Forex trading strategy project, we focused on deriving technical indicators that are widely used in financial markets: Moving Averages and the Relative Strength Index (RSI).

#### **<span style="color:blue">Moving Averages</span>**
Moving averages are used to smooth out price data, making it easier to identify the direction of trends. We calculated two types of moving averages:
- **10-Period Moving Average (MA10)**: This indicator smooths the 'Close' prices over the last 10 periods, helping to identify short-term trends. It is calculated by taking the arithmetic mean of the 'Close' prices for the most recent 10 periods.
- **50-Period Moving Average (MA50)**: This indicator smooths the 'Close' prices over the last 50 periods, helping to identify long-term trends. It is calculated by taking the arithmetic mean of the 'Close' prices for the most recent 50 periods.

These moving averages were plotted against the 'Close' prices to provide a visual representation of short-term and long-term trends in the dataset.

#### **<span style="color:blue">Relative Strength Index (RSI)</span>**
The RSI is a momentum oscillator that measures the speed and change of price movements. It is used to identify overbought or oversold conditions in a market. The RSI values range from 0 to 100, with:
- Readings above 70 indicating that a security is potentially overbought.
- Readings below 30 indicating that a security is potentially oversold.

The calculation of RSI involves:
1. Computing the difference between consecutive 'Close' prices.
2. Separating these differences into gains (positive changes) and losses (negative changes).
3. Averaging the gains and losses over a specified period (14 periods in this project).
4. Calculating the Relative Strength (RS) as the ratio of average gain to average loss.
5. Deriving the RSI using the formula: `RSI = 100 - (100 / (1 + RS))`.

The RSI was then plotted to highlight periods when the market was considered overbought or oversold, providing valuable insights for trading decisions.

Both these features—moving averages and RSI—were saved along with the original dataset to ensure that they could be used in subsequent analysis and strategy implementation. The resulting dataset was stored as a CSV file, ensuring that all newly engineered features were retained for future use.

````python
def calculate_RSI(data, window):
    delta = data.diff()  # Calculating the difference between consecutive data points
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()  # Calculating gains
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()  # Calculating losses
    RS = gain / loss  # Calculating the Relative Strength
    return 100 - (100 / (1 + RS))  # Calculating the RSI
````

````python
forex_data['RSI'] = calculate_RSI(forex_data['Close'], 14)  # Calculating the RSI for 14 periods
````

```python
# Define RSI parameters
rsi_period = 14 # Setting the period for RSI calculation
rsi_buy = 30 # Setting the buy threshold for RSI
rsi_sell = 70 # Setting the sell threshold for RSI

# Calculate RSI
forex_data['RSI'] = ta.momentum.rsi(forex_data['Close'], window=rsi_period) # Calculating RSI for the 'Close' price

# Generate trading signals
forex_data['Signal'] = 0 # Initializing signal column with 0
forex_data['Signal'][forex_data['RSI'] < rsi_buy] = 1 # Setting buy signal
forex_data['Signal'][forex_data['RSI'] > rsi_sell] = -1 # Setting sell signal

# Display the first few rows after generating signals

# Simulate trading
initial_balance = 10000 # Initializing the initial balance
balance = initial_balance # Setting the balance to the initial balance
position = 0 # Initializing position
returns = [] # Initializing returns list

# Iterating over the dataset to apply the trading strategy
for i in range(len(forex_data)):
    if forex_data['Signal'].iloc[i] == 1 and position == 0: # Checking for buy signal
        position = balance / forex_data['Close'].iloc[i] # Calculating position size

    elif forex_data['Signal'].iloc[i] == -1 and position > 0: # Checking for sell signal
