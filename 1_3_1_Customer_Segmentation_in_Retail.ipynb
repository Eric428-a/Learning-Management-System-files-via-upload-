# <span style="color: #1E90FF;">**4. Data Preprocessing**</span>

## <span style="color: #32CD32;">**4.1 Purpose**</span>
Data preprocessing involves cleaning and preparing the dataset for clustering. This includes handling missing values, removing duplicates, and encoding categorical variables.

## <span style="color: #32CD32;">**4.2 Operations**</span>
- **Fill Missing Values**: Apply forward fill to handle any potential missing values.
- **Remove Duplicates**: Eliminate duplicate rows to ensure data integrity.
- **Encode Categorical Variables**: Convert categorical variables into numerical format.

````python
# Data Preprocessing
````

````python
# Fill missing values (if any, though we already know there are none)
````

````python
df.fillna(method='ffill', inplace=True)  # Forward fill missing values (if any) in the DataFrame
````

````python
# Remove duplicates
````

````python
df.drop_duplicates(inplace=True)  # Remove duplicate rows from the DataFrame
````

````python
# Encode the Gender column
````

````python
label_encoder = LabelEncoder()  # Initialize the LabelEncoder
````

````python
df['Gender'] = label_encoder.fit_transform(df['Gender'])  # Encode the Gender column with numerical values
````

````python
# Select features for clustering
````

````python
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]  # Select relevant features for clustering
````

````python
# Normalize the features
````

````python
scaler = StandardScaler()  # Initialize the StandardScaler
````

````python
features_scaled = scaler.fit_transform(features)  # Scale the features to have zero mean and unit variance
````

# <span style="color: #1E90FF;">**5. Clustering (K-Means) and Evaluation**</span>

## <span style="color: #32CD32;">**5.1 Purpose**</span>
This step involves applying the K-Means clustering algorithm to segment the dataset into clusters and evaluating the clustering performance using the silhouette score.

## <span style="color: #32CD32;">**5.2 Operations**</span>
- **Elbow Method**: Determine the optimal number of clusters by plotting WCSS values.
- **Apply K-Means**: Perform clustering using the optimal number of clusters.
- **Evaluate Clustering**: Calculate the silhouette score to assess clustering quality.
- **Visualize Clusters**: Plot the clusters to visualize the segmentation.

````python
# Clustering (K-Means) and Evaluation
````

````python
# Find the optimal number of clusters using the Elbow Method
````

````python
wcss = []  # Initialize an empty list to store WCSS values
````

````python
for i in range(1, 11):  # Iterate over a range of cluster numbers
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)  # Initialize KMeans with i clusters
    kmeans.fit(features_scaled)  # Fit KMeans to the scaled features
    wcss.append(kmeans.inertia_)  # Append the WCSS value to the list
````

````python
# Plot the Elbow Method graph
````

````python
plt.figure(figsize=(10, 6))  # Create a new figure for the Elbow Method plot with specified size
plt.plot(range(1, 11), wcss, marker='o')  # Plot the WCSS values against the number of clusters
plt.title('Elbow Method for Optimal Number of Clusters')  # Set the title of the plot
plt.xlabel('Number of Clusters')  # Set the x-axis label
plt.ylabel('WCSS')  # Set the y-axis label
pdf_pages.savefig()  # Save the current figure to the PDF file
plt.show()  # Display the plot
````

````python
# Apply K-Means clustering
````

````python
optimal_clusters = 5  # Based on the Elbow Method, set the number of clusters to 5
````

````python
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)  # Initialize KMeans with optimal_clusters
````

````python
df['Cluster'] = kmeans.fit_predict(features_scaled)  # Fit KMeans to the scaled features and assign cluster labels to the DataFrame
````

````python
# Evaluate clustering performance using Silhouette Score
````

````python
silhouette_avg = silhouette_score(features_scaled, df['Cluster'])  # Calculate the average silhouette score
````

````python
print(f"\nSilhouette Score for {optimal_clusters} Clusters: {silhouette_avg:.2f}")  # Print the silhouette score
````

## <span style="color: #32CD32;">**6.2 Elbow Method Plot**</span>
- **Description**: The Elbow Method plot helps in determining the optimal number of clusters by plotting the Within-Cluster Sum of Squares (WCSS) against the number of clusters. The "elbow" point on the graph indicates the ideal number of clusters to use.
- **Importance**: Identifying the elbow point helps to choose the number of clusters where adding more clusters does not significantly improve the model's performance.

## <span style="color: #32CD32;">**6.3 Clustering Visualization**</span>
- **Description**: A scatter plot visualizing the clusters created by the K-Means algorithm. The plot displays data points with colors representing different clusters.
- **Features**: The scatter plot uses `Age` and `Annual Income (k$)` as axes, with data points colored based on their assigned cluster.
- **Purpose**: This visualization helps in understanding the distribution and separation of clusters, making it easier to interpret clustering results.

## <span style="color: #32CD32;">**6.4 Silhouette Score**</span>
- **Description**: The silhouette score measures how similar each data point is to its own cluster compared to other clusters. It ranges from -1 to +1, where a higher score indicates better-defined clusters.
- **Importance**: Provides an evaluation of clustering quality, helping to understand how well-separated the clusters are.
