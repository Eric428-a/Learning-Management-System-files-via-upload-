## Step 3: Data Preprocessing

### Handling Missing Values

- **<font color='blue'>Identifying Missing Values</font>**:
  - **Approaches**: Check for missing values represented as '?', NaNs, or empty strings.
  - **Tools**: Utilize pandas functions like `pd.isnull()` and `pd.notnull()` to detect missing values. Visualization techniques can also help identify patterns of missing data.

### Imputation Strategies

- **<font color='orange'>Strategies for Imputation</font>**:
  - **Mean Imputation**: Replace missing values with the mean of the feature. Suitable for numerical data with a symmetric distribution.
  - **Median Imputation**: Replace missing values with the median of the feature. Useful when the data contains outliers.
  - **Mode Imputation**: Replace missing categorical values with the mode (most frequent value). Ideal for categorical variables.

### Data Transformation and Scaling

- **<font color='green'>Feature Scaling</font>**:
  - **Standardization**: Scale features to have zero mean and unit variance. Helps in dealing with features that have different scales.
  - **Normalization**: Scale features to a specified range, typically [0, 1]. Useful for algorithms that require features to be on a similar scale.

### Exploratory Data Analysis (EDA)

- **<font color='purple'>Statistical Summaries</font>**:
  - Calculate descriptive statistics such as mean, median, min, max, and quartiles for numeric features. Provides insights into central tendency and spread of data.

### Visualization for Insights

- **<font color='red'>Data Distribution Plots</font>**:
  - Utilize histograms and density plots to visualize the distribution of numeric features. Helps understand the shape, skewness, and presence of outliers in data.

### Feature Engineering

- **<font color='orange'>Creating New Features</font>**:
  - Generate new features from existing ones to capture additional patterns and improve predictive performance of models. Example: creating interaction terms or polynomial features.

### Handling Categorical Variables

- **<font color='purple'>Encoding Categorical Data</font>**:
  - Convert categorical variables into numerical representations suitable for machine learning algorithms. Common methods include one-hot encoding and label encoding.

```python
# Handle missing values represented by '?'
```

```python
df.replace('?', pd.NA, inplace=True)  # Replace '?' with NA (missing values)
```

```python
# Convert columns to numeric
```

```python
for col in ['ca', 'thal']:
    df[col] = pd.to_numeric(df[col])  # Convert 'ca' and 'thal' columns to numeric
```

```python
# Impute missing values
```

```python
imputer = SimpleImputer(strategy='mean')  # Create an imputer to fill missing values with the mean
```

```python
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)  # Apply the imputer and create a new DataFrame
```

```python
# Assuming df_imputed is already defined with appropriate preprocessing steps
```

```python
# Correlation heatmap
```

```python
plt.figure(figsize=(12, 8))  # Set the figure size for the heatmap
sns.heatmap(df_imputed.corr(), annot=True, cmap='coolwarm', fmt=".2f")  # Create heatmap with correlation matrix
plt.title('Correlation Heatmap')  # Set title for the plot
plt.savefig('correlation_heatmap.pdf')  # Save the correlation heatmap as PDF
plt.show()  # Display the heatmap
```

```python
# Distribution plots for numeric variables
```

```python
numeric_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']  # Define list of numeric columns for plotting
```

```python
plt.figure(figsize=(15, 10))  # Set the overall figure size for distribution plots
for i, col in enumerate(numeric_cols, start=1):  # Iterate over numeric columns
    plt.subplot(2, 3, i)  # Create subplots in a 2x3 grid
    sns.histplot(df_imputed[col], kde=True, bins=20, color='skyblue')  # Plot histogram with kernel density estimation
    plt.title(f'Distribution of {col}')  # Set title for each subplot
    plt.xlabel(col)  # Set x-axis label
    plt.ylabel('Count')  # Set y-axis label
plt.tight_layout()  # Adjust subplot parameters to give specified padding
plt.savefig('numeric_distributions.pdf')  # Save numeric distributions as PDF
plt.show()  # Display the plots
```

```python
# Boxplot for age vs target (num)
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size for the boxplot
sns.boxplot(x='num', y='age', data=df_imputed, palette='Set2')  # Create boxplot for age vs heart disease status
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])  # Set x-axis tick labels
plt.title('Age Distribution by Heart Disease Status')  # Set title for the plot
plt.xlabel('Heart Disease Status')  # Set x-axis label
plt.ylabel('Age')  # Set y-axis label
plt.savefig('age_vs_heart_disease.pdf')  # Save age vs heart disease status boxplot as PDF
plt.show()  # Display the plot
```

```python
# Convert target to binary classification
```

```python
y = df_imputed['num'].apply(lambda x: 1 if x > 0 else 0)  # Convert the target variable into binary (1 if disease present, 0 otherwise)
```

```python
# Split the data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)  # Split the data into training and testing sets
```

```python
# Initialize the model
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  # Create a Random Forest classifier
```

```python
# Train the model
```

```python
rf_model.fit(X_train, y_train)  # Train the model on the training data
```

```python
# Make predictions
```

```python
y_pred = rf_model.predict(X_test)  # Make predictions on the testing data
```

```python
# Save predictions to CSV
```

```python
predictions_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  # Create a DataFrame for actual and predicted values
```

```python
predictions_df.to_csv('predictions.csv', index=False)  # Save the predictions to a CSV file
```

```python
# Evaluate the model
```

```python
print("\nAccuracy:", accuracy_score(y_test, y_pred))  # Print the accuracy of the model
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))  # Print the confusion matrix
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))  # Print the classification report
```

```python
# Plot confusion matrix
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)  # Create a confusion matrix
plt.figure(figsize=(10,7))  # Create a new figure for the plot
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')  # Plot the confusion matrix
plt.xlabel('Predicted')  # Label the x-axis
plt.ylabel('Actual')  # Label the y-axis
plt.title('Confusion Matrix')  # Title of the plot
plt.savefig('confusion_matrix.pdf')  # Save the plot as a PDF file
plt.show()  # Display the plot
```

```python
# Feature importance
```

```python
importances = rf_model.feature_importances_  # Get feature importances from the model
```

```python
feature_names = df_imputed.drop(columns='num').columns  # Get the feature names
```

```python
feat_importances = pd.Series(importances, index=feature_names)  # Create a Series with feature importances
```

```python
plt.figure(figsize=(10,7))  # Create a new figure for the plot
feat_importances.nlargest(10).plot(kind='barh')  # Plot the top 10 important features
plt.xlabel('Feature Importance')  # Label the x-axis
plt.ylabel('Features')  # Label the y-axis
plt.title('Top 10 Important Features')  # Title of the plot
plt.savefig('feature_importance.pdf')  # Save the plot as a PDF file
plt.show()  # Display the plot
```
