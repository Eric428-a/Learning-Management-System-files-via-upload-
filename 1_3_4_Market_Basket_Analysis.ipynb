```python
def check_special_characters(df):
    """
    Checks for special characters in string columns of a DataFrame.

    Args:
    df (pd.DataFrame): Input DataFrame.

    Returns:
    set: Set of special characters found in string columns.
    """
    special_characters = set()  # Initializing an empty set for special characters
    for column in df.select_dtypes(include=[object]).columns:
        for value in df[column].dropna().unique():
            special_characters.update([char for char in str(value) if not char.isalnum() and not char.isspace()])
    return special_characters
```
