### <span style="color:blue">**Step 7: Scale Numerical Features**</span>

#### <span style="color:green">**Notes:**</span>

1. **<span style="color:#880015">Why Scale Numerical Features?</span>**
   - Scaling numerical features ensures that all features contribute equally to the model fitting process.
   - Machine learning algorithms like Logistic Regression and SVMs often perform better when features are on a similar scale.
   - It prevents features with larger ranges from dominating the learning process.

2. **<span style="color:#880015">StandardScaler:</span>**
   - Use `StandardScaler` from scikit-learn to standardize numerical features.
   - `StandardScaler` transforms the data to have a mean of 0 and a standard deviation of 1.
   - It works well when the data follows a normal distribution.

3. **<span style="color:#880015">Scaling Process:</span>**
   - Fit `StandardScaler` on the training data (`X_train`) to learn the parameters (mean and standard deviation) needed for scaling.
   - Transform both the training set (`X_train`) and the testing set (`X_test`) using the fitted scaler.

4. **<span style="color:#880015">Preserving Data Integrity:</span>**
   - Save the scaled data (`X_train_scaled` and `X_test_scaled`) to CSV files for future model training and evaluation.
   - This ensures consistency in preprocessing steps across different runs of the model.

5. **<span style="color:#880015">Considerations:</span>**
   - Scaling should be performed after splitting the data into training and testing sets to avoid data leakage.
   - Not all algorithms require scaling (e.g., Tree-based models), but it generally improves convergence and performance for many.

Scaling numerical features is a critical preprocessing step that enhances the effectiveness of machine learning models, ensuring robust and reliable predictions.

```python
# Scale numerical features
```

```python
scaler = StandardScaler()  # Initializing the standard scaler
```

```python
X_train = scaler.fit_transform(X_train)  # Fitting and transforming the training features
```

```python
X_test = scaler.transform(X_test)  # Transforming the testing features
```

```python
# Save the scaled data to CSV
```

```python
pd.DataFrame(X_train).to_csv('X_train_scaled.csv', index=False)  # Saving the scaled training features to CSV
```

```python
pd.DataFrame(X_test).to_csv('X_test_scaled.csv', index=False)  # Saving the scaled testing features to CSV
```

```python
# Model Building
```

```python
# Logistic Regression
```

```python
log_reg = LogisticRegression(max_iter=1000)  # Initializing the logistic regression model with 1000 max iterations
```

```python
log_reg.fit(X_train, y_train)  # Fitting the logistic regression model to the training data
```

```python
y_pred_log_reg = log_reg.predict(X_test)  # Making predictions on the testing data using the logistic regression model
```

```python
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)  # Calculating the accuracy of the logistic regression model
```

```python
print("\nLogistic Regression accuracy:")  # Printing a message
print(accuracy_log_reg)  # Displaying the accuracy of the logistic regression model
```

```python
print(classification_report(y_test, y_pred_log_reg))  # Displaying the classification report for the logistic regression model
```

```python
print(confusion_matrix(y_test, y_pred_log_reg))  # Displaying the confusion matrix for the logistic regression model
```

```python
# Random Forest
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)  # Initializing the random forest model with 100 estimators
```

```python
rf.fit(X_train, y_train)  # Fitting the random forest model to the training data
```

```python
y_pred_rf = rf.predict(X_test)  # Making predictions on the testing data using the random forest model
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)  # Calculating the accuracy of the random forest model
```

```python
print("\nRandom Forest accuracy:")  # Printing a message
print(accuracy_rf)  # Displaying the accuracy of the random forest model
```

```python
print(classification_report(y_test, y_pred_rf))  # Displaying the classification report for the random forest model
```

```python
print(confusion_matrix(y_test, y_pred_rf))  # Displaying the confusion matrix for the random forest model
```
