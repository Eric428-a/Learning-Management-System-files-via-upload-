The Mean Reversion Strategy is a financial trading strategy that capitalizes on the tendency of asset prices to revert to their historical mean or average over time. This project aims to develop and implement a mean reversion trading strategy using machine learning and statistical techniques. Below is a comprehensive overview of the project, including every aspect of machine learning and financial analysis related to this strategy.

### <font color='darkorange'>**1. Objective**</font>

- **Goal**: The primary goal of this strategy is to exploit the assumption that asset prices, which deviate significantly from their historical average, are likely to revert back to that average. This strategy involves identifying such deviations and making trading decisions based on them.

### <font color='darkorange'>**2. Data Collection and Preparation**</font>

- **Data Collection**:
  - **Source**: Historical price data of assets (e.g., stocks, currencies) is collected. This data is typically sourced from financial databases or APIs like Yahoo Finance, Alpha Vantage, or Quandl.
  - **Data Fields**: Important fields include Date, Open, High, Low, Close, and Volume.

- **Data Preparation**:
  - **Load Data**: Read the historical price data into a DataFrame.
    ```python
    import pandas as pd
    data = pd.read_csv('path/to/historical_prices.csv')  # Load historical price data
    ```

  - **Data Cleaning**: Handle missing values and ensure data consistency.
    ```python
    data.dropna(inplace=True)  # Remove rows with missing values
    ```

### <font color='darkorange'>**3. Feature Engineering**</font>

- **Calculate Moving Average**:
  - Compute the moving average (e.g., 20-day, 50-day) of the asset prices to represent the mean price over time.
    ```python
    data['Moving_Avg_20'] = data['Close'].rolling(window=20).mean()  # 20-day moving average
    ```

- **Calculate Z-Score**:
  - Normalize the price data to identify deviations from the moving average using the Z-score.
    ```python
    data['Mean'] = data['Close'].rolling(window=20).mean()  # Calculate rolling mean
    data['Std'] = data['Close'].rolling(window=20).std()  # Calculate rolling standard deviation
    data['Z_Score'] = (data['Close'] - data['Mean']) / data['Std']  # Calculate Z-score
    ```

### <font color='darkorange'>**4. Strategy Implementation**</font>

- **Define Trading Signals**:
  - Establish trading signals based on the Z-score. For example, buy signals when the Z-score is below a certain threshold (e.g., -2) and sell signals when the Z-score is above a certain threshold (e.g., 2).
    ```python
    data['Signal'] = 0  # Initialize signals
    data.loc[data['Z_Score'] < -2, 'Signal'] = 1  # Buy signal
    data.loc[data['Z_Score'] > 2, 'Signal'] = -1  # Sell signal
    ```

- **Backtesting**:
  - Test the strategy using historical data to evaluate its performance and profitability.
    ```python
    data['Returns'] = data['Close'].pct_change()  # Calculate daily returns
    data['Strategy_Returns'] = data['Signal'].shift(1) * data['Returns']  # Strategy returns
    ```

### <font color='darkorange'>**5. Performance Evaluation**</font>

- **Calculate Performance Metrics**:
  - Evaluate the performance of the strategy using metrics such as cumulative returns, Sharpe ratio, and maximum drawdown.
    ```python
    cumulative_returns = (1 + data['Strategy_Returns']).cumprod() - 1  # Cumulative returns
    sharpe_ratio = data['Strategy_Returns'].mean() / data['Strategy_Returns'].std() * (252 ** 0.5)  # Sharpe ratio
    ```

- **Visualize Results**:
  - Plot the performance metrics and trading signals to visualize the effectiveness of the strategy.
    ```python
    import matplotlib.pyplot as plt

    plt.xlabel('Date')
    plt.ylabel('Cumulative Returns')
    ```

# <font color='blue'>**Implement Mean Reversion Strategy**</font>

## <font color='darkgreen'>**Objective**</font>

The objective of this step is to apply the mean reversion strategy using Z-scores to identify buy and sell opportunities.

### <font color='darkorange'>**1. Calculate Z-Score for Mean Reversion**</font>

#### **a. Calculate Z-Score**

- **Purpose**: To normalize prices and generate mean reversion signals.
- **Example Code**:
    ```python
    # Calculate 20-day rolling mean and standard deviation
    df['mean'] = df['close'].rolling(window=20, min_periods=1).mean()
    df['std'] = df['close'].rolling(window=20, min_periods=1).std()
    
    # Calculate Z-score
    df['z_score'] = (df['close'] - df['mean']) / df['std']
    ```

### <font color='darkorange'>**2. Generate Mean Reversion Signals**</font>

#### **a. Create Buy/Sell Signals Based on Z-Score**

- **Purpose**: To generate buy and sell signals based on the Z-score thresholds.
- **Example Code**:
    ```python
    # Initialize the mean reversion signal column with zeros
    df['mean_reversion_signal'] = 0
    
    # Set signal to -1 for sell when Z-score > 1
    df.loc[df['z_score'] > 1, 'mean_reversion_signal'] = -1
    
    # Set signal to 1 for buy when Z-score < -1
    df.loc[df['z_score'] < -1, 'mean_reversion_signal'] = 1
    ```

## <font color='darkorange'>**3. Visualize Z-Score**</font>

### **a. Plot Z-Score and Thresholds**

- **Purpose**: To visualize the Z-score and its thresholds for mean reversion strategy analysis.
- **Example Code**:
    ```python
    # Plot Z-score
    plt.figure(figsize=(14, 7))
    plt.plot(df['date'], df['z_score'], label='Z-Score')
    plt.axhline(1, color='r', linestyle='--', label='Sell Threshold')
    plt.axhline(-1, color='g', linestyle='--', label='Buy Threshold')
    plt.xlabel('Date')
    plt.ylabel('Z-Score')
    plt.title('Z-Score for Mean Reversion Strategy')
    plt.legend()
    plt.savefig('/content/z_score_mean_reversion.pdf')  # Save the plot to a PDF file
    plt.show()  # Display the plot
    ```

````python
# Mean Reversion Strategy
````

````python
# Calculate Z-score for normalization
````

````python
df['mean'] = df['close'].rolling(window=20, min_periods=1).mean()  # Calculating 20-day rolling mean
````

````python
df['std'] = df['close'].rolling(window=20, min_periods=1).std()  # Calculating 20-day rolling standard deviation
````

````python
df['z_score'] = (df['close'] - df['mean']) / df['std']  # Calculating Z-score
````

````python
# Create signals based on Z-score
````

````python
df['mean_reversion_signal'] = 0  # Initializing the mean reversion signal column with zeros
````

````python
df.loc[df['z_score'] > 1, 'mean_reversion_signal'] = -1  # Setting signal to -1 for sell when Z-score > 1
````

````python
df.loc[df['z_score'] < -1, 'mean_reversion_signal'] = 1  # Setting signal to 1 for buy when Z-score < -1
````

````python
# Plot Z-score
````

````python
plt.figure(figsize=(14, 7))  # Setting the figure size for the plot
plt.plot(df['date'], df['z_score'], label='Z-Score')  # Plotting Z-score
plt.axhline(1, color='r', linestyle='--', label='Sell Threshold')  # Adding a sell threshold line at Z-score = 1
plt.axhline(-1, color='g', linestyle='--', label='Buy Threshold')  # Adding a buy threshold line at Z-score = -1
plt.xlabel('Date')  # Setting x-axis label
plt.ylabel('Z-Score')  # Setting y-axis label
plt.title('Z-Score for Mean Reversion Strategy')  # Setting the title of the plot
plt.legend()  # Adding a legend to the plot
plt.savefig('/content/z_score_mean_reversion.pdf')  # Saving the plot to a PDF file
plt.show()  # Displaying the plot
````
