```python
url = "https://www.gutenberg.org/files/1342/1342-0.txt"  # URL of the book to scrape
```

```python
# Function to scrape the text from the Project Gutenberg URL
```

```python
def scrape_gutenberg_text(url):
    response = requests.get(url)  # Make a GET request to the URL
    if response.status_code == 200:  # Check if the request was successful
        text = response.text  # Get the text content from the response
        print(f"Successfully scraped the text from {url}")  # Print success message
        return text  # Return the scraped text
    else:
        print(f"Failed to retrieve the book. Status code: {response.status_code}")  # Print error message if request failed
        return None  # Return None if scraping failed
```

```python
# Scrape the text from the provided URL
```

```python
book_text = scrape_gutenberg_text(url)  # Call the scrape function
```

```python
# Display first 500 characters of the text to ensure proper scraping
```

```python
if book_text:  # Check if text was successfully scraped
    print("\nPreview of scraped text (first 500 characters):")  # Print preview header
    print(book_text[:500])  # Print the first 500 characters of the scraped text
```

```python
def load_and_preview_text(file_path):
    # Open and read the file
    with open(file_path, 'r', encoding='utf-8') as file:  # Open the file in read mode with UTF-8 encoding
        data = file.readlines()  # Read all lines from the file

    # Display the first 10 lines for a quick preview
    preview = "".join(data[:10])  # Join the first 10 lines into a single string
    print("File Preview:\n", preview)  # Print the preview
    return data  # Return the list of lines from the file
```

```python
if os.path.exists(file_path):  # Check if the file exists
    # Open and read the file
    with open(file_path, 'r', encoding='utf-8') as file:  # Open the file in read mode with UTF-8 encoding
        scraped_text = file.read()  # Read the entire text into a single string

    # Display basic file details
    print("File loaded successfully.")  # Print success message
    print(f"Total length of text: {len(scraped_text)} characters\n")  # Print the total character count

    # Preview first 500 characters of the text
    print("Preview of the scraped text (first 500 characters):")  # Print preview header
    print(scraped_text[:500])  # Print the first 500 characters of the scraped text

    # Function to clean text for word analysis
    def clean_text_for_analysis(text):
        # Convert to lowercase
        text = text.lower()  # Convert the text to lowercase
        # Remove non-alphabetical characters
        text = re.sub(r'[^a-z\s]', '', text)  # Remove all non-alphabetic characters using regex
        # Remove multiple spaces and newlines
        text = re.sub(r'\s+', ' ', text).strip()  # Replace multiple spaces with a single space and strip whitespace
        return text  # Return the cleaned text

    # Clean the scraped text for word analysis
    cleaned_text = clean_text_for_analysis(scraped_text)  # Call the cleaning function

    # Split into words
    words = cleaned_text.split()  # Split the cleaned text into individual words

    # Total word count
    total_word_count = len(words)  # Count total words

    # Unique word count
    unique_words = set(words)  # Create a set of unique words
    unique_word_count = len(unique_words)  # Count unique words

    # Top 10 most frequent words
    word_frequencies = Counter(words)  # Count the frequency of each word
    top_10_words = word_frequencies.most_common(10)  # Get the 10 most common words

    # Display text analysis details
    print(f"\nTotal word count: {total_word_count}")  # Print total word count
    print(f"Unique word count: {unique_word_count}")  # Print unique word count
    print(f"\nTop 10 most frequent words: {top_10_words}")  # Print top 10 frequent words

else:
    print(f"File not found at path: {file_path}")  # Print error message if the file is not found
```

```python
import streamlit as st  # Import the Streamlit library for creating web apps
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel  # Import the GPT-2 tokenizer and model from Hugging Face
```

```python
import torch  # Import PyTorch for tensor operations
```

```python
import nltk  # Import the Natural Language Toolkit for text processing
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer  # Import the SentimentIntensityAnalyzer for sentiment scoring
```

```python
# Load the model and tokenizer (make sure the model is downloaded or available)
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')  # Load the GPT-2 tokenizer
```

```python
model = GPT2LMHeadModel.from_pretrained('gpt2')  # Load the GPT-2 language model
```

```python
nltk.download('vader_lexicon')  # Download the VADER lexicon for sentiment analysis
```

```python
sia = SentimentIntensityAnalyzer()  # Initialize the sentiment intensity analyzer
```

```python
def analyze_sentiment(text):  # Define a function to analyze sentiment
    sentiment_scores = sia.polarity_scores(text)  # Get sentiment scores for the input text
    return sentiment_scores  # Return the sentiment scores
```

```python
st.title("Text Analysis App")  # Set the title of the Streamlit app
```

```python
st.write("Enter text for sentiment analysis:")  # Write instructions for the user
```

```python
user_input = st.text_area("Input Text", "")  # Create a text area for user input
```

```python
if st.button("Analyze"):  # Create a button to trigger sentiment analysis
    if user_input:  # Check if the user has entered text
        sentiment = analyze_sentiment(user_input)  # Analyze the sentiment of the input text
        st.write("Sentiment Scores:")  # Display the sentiment scores label
        st.json(sentiment)  # Display the sentiment scores in JSON format
    else:  # If no text is entered
        st.error("Please enter some text.")  # Show an error message
```
