```python
# We will use IQR method to detect and remove outliers
```

```python
def remove_outliers(df, column):  # Define a function to remove outliers using IQR method
    Q1 = df[column].quantile(0.25)  # Calculate the first quartile (25th percentile) of the column
    Q3 = df[column].quantile(0.75)  # Calculate the third quartile (75th percentile) of the column
    IQR = Q3 - Q1  # Calculate the Interquartile Range (IQR)
    lower_bound = Q1 - 1.5 * IQR  # Calculate the lower bound for outlier detection
    upper_bound = Q3 + 1.5 * IQR  # Calculate the upper bound for outlier detection
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]  # Return rows where column values are within bounds
```

```python
columns_to_check = ['High', 'Low', 'Open', 'Close', 'Volume', 'Marketcap']  # List of columns to check for outliers
```

```python
for column in columns_to_check:  # Loop through each column
    data = remove_outliers(data, column)  # Remove outliers from the column
```

  - **Apply Signal Generation**: Compute signals for each period and add them as new columns in the dataset.
  - **Result**: The dataset now includes columns representing buy and sell signals based on the defined criteria.

  - **Plot Signals**: Create a line plot showing 'Close' prices along with buy and sell signals for a selected period (e.g., 90 days).

## <span style="color:#1f77b4">8. Calculating MACD (Moving Average Convergence Divergence)</span>

### <span style="color:#ff7f0e">8.1. Introduction to MACD</span>

- **Objective**: Calculate the MACD (Moving Average Convergence Divergence) to further enhance trading signals.
- **Description**: The MACD is a trend-following momentum indicator that shows the relationship between two moving averages of a security's price. It consists of:
  - **MACD Line**: Difference between the 12-day EMA (Exponential Moving Average) and the 26-day EMA.
  - **Signal Line**: 9-day EMA of the MACD Line.
  - **Histogram**: Difference between the MACD Line and the Signal Line.

### <span style="color:#2ca02c">8.2. Calculating MACD Components</span>

- **Objective**: Compute the MACD line and Signal line to analyze trend and momentum.
- **Actions**:
  - **Calculate EMAs**: Compute the 12-day and 26-day EMAs of the 'Close' price.
    - **12-day EMA**: `data['EMA_12'] = data['Close'].ewm(span=12, adjust=False).mean()`
    - **26-day EMA**: `data['EMA_26'] = data['Close'].ewm(span=26, adjust=False).mean()`
  - **Compute MACD Line**: Subtract the 26-day EMA from the 12-day EMA.
    - **Formula**: `data['MACD'] = data['EMA_12'] - data['EMA_26']`
  - **Calculate Signal Line**: Compute the 9-day EMA of the MACD Line.
    - **Formula**: `data['Signal'] = data['MACD'].ewm(span=9, adjust=False).mean()`

### <span style="color:#d62728">8.3. Plotting MACD and Signal Line</span>

- **Objective**: Visualize the MACD Line and Signal Line to identify potential trading signals.
- **Actions**:
  - **Plot Lines**: Create a line plot to show the MACD Line and Signal Line over time.
    - **MACD Line**: Plotted in blue.
    - **Signal Line**: Plotted in red.
  - **Title and Labels**: Add titles and labels to the plot for clarity.
    - **Title**: "MACD and Signal Line"
    - **X-axis**: 'Date'
    - **Y-axis**: 'Value'
  - **Interpretation**: Analyze the plot to identify crossovers and trends:
    - **Buy Signal**: When the MACD Line crosses above the Signal Line.
    - **Sell Signal**: When the MACD Line crosses below the Signal Line.

```python
# Calculate MACD
```

```python
data['EMA_12'] = data['Close'].ewm(span=12, adjust=False).mean()  # Calculate the 12-day EMA
```

```python
data['EMA_26'] = data['Close'].ewm(span=26, adjust=False).mean()  # Calculate the 26-day EMA
```

```python
data['MACD'] = data['EMA_12'] - data['EMA_26']  # Calculate MACD line
```

```python
data['Signal'] = data['MACD'].ewm(span=9, adjust=False).mean()  # Calculate Signal line
```

```python
# Plot MACD and Signal Line
```

```python
plt.figure(figsize=(14, 7))  # Set figure size for the plot
plt.plot(data['Date'], data['MACD'], label='MACD', color='blue')  # Plot MACD line
plt.plot(data['Date'], data['Signal'], label='Signal Line', color='red')  # Plot Signal line
plt.title('MACD and Signal Line')  # Set the title of the plot
plt.xlabel('Date')  # Set the x-axis label
plt.ylabel('Value')  # Set the y-axis label
plt.legend()  # Show legend
plt.show()  # Display the plot
```
